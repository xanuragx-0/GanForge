```python
# Importing required libraries
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Reading image using OpenCV
img = cv2.imread("0001.jpg")
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# Displaying image as a Numpy Array and subsequently converting it into a Pandas series
img.shape
pd.Series(img.flatten()).value_counts()
```

```python
# Illustrating difference between OpenCV (BGR) and Matplotlib (RGB) reading the image
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
img1=cv2.imread("0001.jpg")
img2=plt.imread("0001.jpg")
axs[0].imshow(img1)
axs[1].imshow(img2)
axs[0].axis("off")
axs[1].axis("off")
axs[0].set_title("cv2: BGR")
axs[1].set_title("plt: RGB")
plt.show()
```

```python
# Converting BGR to RGB
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
# Displaying the image in the three separate colour channels
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(img[:, :, 0], cmap="Reds")
axs[1].imshow(img[:, :, 1], cmap="Greens")
axs[2].imshow(img[:, :, 2], cmap="Blues")
for ax in axs:
    ax.axis("off")
plt.show()
```

```python
# Converting to GrayScale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.colorbar()
plt.show()
```

```python
#Converting to Thermal
heat = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(heat, cmap="plasma")
plt.axis("off")
plt.colorbar()
plt.show()
```

```python
# Resizing
img_resized = cv2.resize(img, None, fx=0.1, fy=0.1)
plt.imshow(img_resized)
plt.axis("off")
plt.show()
```

```python
# Resizing in two dimensions
img_resized1 = cv2.resize(img, (100,200))
img_resized2 = cv2.resize(img, (200,100))
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_resized1)
axs[1].imshow(img_resized2)
axs[0].axis("off")
axs[1].axis("off")
plt.show()
```

```python
# Filtering - Image Sharpening
img_sharpened = cv2.filter2D(img, -1, np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]]))
plt.imshow(img_sharpened)
plt.axis("off")
plt.show()
```

```python
# Filtering - Image Blurring
img_blurred = cv2.filter2D(img, -1, np.ones((3,3), np.float32)/9)
plt.imshow(img_blurred)
plt.axis("off")
plt.show()
```
