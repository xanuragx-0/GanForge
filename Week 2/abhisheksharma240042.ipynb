```python
#importing important libraries
from google.colab import files
uploaded = files.upload()  # Choose the file manually

import numpy as np

import cv2
import matplotlib.pylab as plt
```

```python

img=cv2.imread('carphoto.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.title("original image")
plt.show()
```

```python
#converting colorful to gray scale image
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
#applying gaussian filter to make it smooth
blur = cv2.GaussianBlur(gray, (5, 5), 0)

plt.imshow(blur, cmap='gray')
plt.title("Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
#applying canny filter to detect the edges

edges = cv2.Canny(blur, 50, 150)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```

```python
#applying thresholding
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
#resizing the image
resized = cv2.resize(img, (300, 300))
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(resized_rgb)
plt.title("Resized to 300x300")
plt.axis('off')
plt.show()
```

```python
#  Contour Detection
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
img_contours = img.copy()
cv2.drawContours(img_contours, contours,-1, (0, 255, 0), 2)  # green contours7
img_contours_rgb = cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB)
plt.imshow(img_contours_rgb)
plt.title(f"Contours Found: {len(contours)}")
plt.axis('off')
plt.show()
```

```python
#seperating image into RED GREEN AND BLUE channel

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


R, G, B = cv2.split(img_rgb)

    # Create subplots
plt.figure(figsize=(15, 5))

    # Red channel
plt.subplot(1, 3, 1)
plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

    # Green channel
plt.subplot(1, 3, 2)
plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

    # Blue channel
plt.subplot(1, 3, 3)
plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

    # Show all plots
plt.tight_layout()
plt.show()
```

```python
# show cv filters along with kernel

# Sharpening Kernel
sharpen_kernel = np.array([[-1, -1, -1],
                           [-1,  9, -1],
                           [-1, -1, -1]])
sharpened_image = cv2.filter2D(img, -1, sharpen_kernel)
sharpened_image_rgb = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)
plt.imshow(sharpened_image_rgb)
plt.title("Sharpened Image")
plt.axis('off')
plt.show()



```

```python
# Emboss Kernel
emboss_kernel = np.array([[-2, -1,  0],
                          [-1,  1,  1],
                          [ 0,  1,  2]])
embossed_image = cv2.filter2D(img, -1, emboss_kernel)
embossed_image_rgb = cv2.cvtColor(embossed_image, cv2.COLOR_BGR2RGB)
plt.imshow(embossed_image_rgb)
plt.title("Embossed Image")
plt.axis('off')
plt.show()
```
