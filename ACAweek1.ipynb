```python
import pandas as pd
import numpy as np

from glob import glob

import cv2
import matplotlib.pylab as plt
```

```python
cat_files = glob('/content/images (1).jpeg')
dog_files = glob('/content/images.jpeg')
```

```python
img_mpl= plt.imread(cat_files[0])
img_cv2 = cv2.imread(cat_files[0])
img_mpl.shape, img_cv2.shape
```

```python
pd.Series(img_mpl.flatten()).plot(kind='hist', bins=50, title="Distribution of Pixel Values")
plt.show()
```

```python
fig, ax= plt.subplots(figsize=(10,10))
ax.imshow(img_mpl)
ax.axis('off')
plt.show()
```

```python
fig, axs=plt.subplots(1,3,figsize=(15,5))
axs[0].imshow(img_mpl[:, :,0],cmap='Reds')
axs[1].imshow(img_mpl[:, :,1],cmap='Greens')
axs[2].imshow(img_mpl[:, :,2],cmap='Blues')
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red Channel')
axs[1].set_title('Green Channel')
axs[2].set_title('Blue Channel')
plt.show()
```

```python
fig, axs=plt.subplots(1,2,figsize=(10,5))
axs[0].imshow(img_cv2)
axs[1].imshow(img_mpl)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('CV2 Image')
axs[1].set_title('Matplotlib Image')
plt.show()
```

```python
ig_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
fig, axs=plt.subplots()
axs.imshow(ig_cv2_rgb)
axs.axis('off')
plt.show()
```

```python
img=plt.imread(dog_files[0])
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(img)
axs.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(img_gray, cmap='Greys')
axs.axis('off')
axs.set_title('Gray Image')
plt.show()
```

```python
img_resized= cv2.resize(img, None, fx=0.25,fy=0.25)
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resized)
axs.axis('off')
plt.show()
```

```python
img_resize= cv2.resize(img, (100,200))
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resize)
axs.axis('off')
plt.show()
```

```python
img_resize= cv2.resize(img, (5000,5000), interpolation= cv2.INTER_CUBIC)
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resize)
axs.axis('off')
plt.show()
```

```python
kernel_sharpening=np.array([[-1,-1,-1],
                           [-1,9,-1],
                           [-1,-1,-1]])
sharpened= cv2.filter2D(img, -1, kernel_sharpening)
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(sharpened)
axs.axis('off')
axs.set_title('Sharpened Image')
plt.show()
```

```python
kernel_3x3=np.ones((3,3), dtype=np.float32)/9
blurred= cv2.filter2D(img, -1, kernel_3x3)
fig, axs=plt.subplots(figsize=(8,8))
axs.imshow(blurred)
axs.axis('off')
axs.set_title('Blurred Image')
plt.show()
```

```python
plt.imsave('mpl_dog.png',blurred)
cv2.imwrite('cv2_dog.png',blurred)
```
