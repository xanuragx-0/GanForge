## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
plt.figure()
ax1 = plt.axes()
ax2 = plt.axes([0.2,0.5,.2,.2])
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Create figure
fig = plt.figure()

# Add axes
ax1 = fig.add_axes([0, 0, 1, 1])           # Main axes
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])   # Inset axes

# Plot on both axes
ax1.plot(x, y)
ax2.plot(x, y)

# Show the figure
plt.show()
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])           # Main plot
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])   # Inset plot

# Plot on main axes
ax1.plot(x, z, 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('Main Plot')
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Main plot
ax1.plot(x, y, 'r-')
ax1.set_title('Main Plot')

# Zoomed plot
ax2.plot(x, y, 'b')
ax2.set_xlim(20, 40)
ax2.set_ylim(50, 200)
ax2.set_title('Zoomed View')

plt.tight_layout()
plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
plt.subplots(nrows=1, ncols=2)
figsize=(12, 4)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
#using linewidth and style inside the plot function to modify the plot

# Create figure with two subplots and custom size
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# First subplot (x,y) with dashed line and increased linewidth
ax1.plot(x, y, linestyle='--', linewidth=3, color='blue')
ax1.set_title('Plot of (x,y)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Second subplot (x,z) with dotted line and different color
ax2.plot(x, z, linestyle=':', linewidth=2, color='red')
ax2.set_title('Plot of (x,z)')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
#changing the the lenght and width of the figure using the figsize funtion

# Create figure with two subplots and custom size (width, height)
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))  # Wider figure

# First subplot
ax1.plot(x, y, linestyle='--', linewidth=3, color='blue')
ax1.set_title('Plot of (x,y)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Second subplot
ax2.plot(x, z, linestyle=':', linewidth=2, color='red')
ax2.set_title('Plot of (x,z)')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

plt.tight_layout()
plt.show()
```
