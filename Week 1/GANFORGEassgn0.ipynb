```python
import matplotlib.pyplot as plt

print("Exercise 1")

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
xpoints = [0, 100]
ypoints = [0, 200]
ax.plot(xpoints, ypoints)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

print("Exercise 3")

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.set_xlabel("x")
ax1.set_ylabel("z")

ax2.set_xlabel("x")
ax2.set_ylabel("y")

xpoints1 = np.linspace(0, 100, 1000)
xpoints1_array = np.array(xpoints1)
ypoints1 = xpoints1_array ** 2
ax1.plot(xpoints1, ypoints1, 'r')

xpoints2 = np.linspace(20, 22, 500)
ypoints2 = xpoints2 * 2
ax2.plot(xpoints2, ypoints2, 'r')

plt.show()

```
